## Alpha Vantage

```python
# Replace the "demo" API key below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=RELIANCE.BSE&outputsize=full&apikey=YOUR_API_KEY'
r = requests.get(url)
data = r.json()

# Extract the daily time series data
daily_data = data['Time Series (Daily)']

# Create a DataFrame from the JSON data
df = pd.DataFrame.from_dict(daily_data, orient='index')

# Reset the index and rename columns
df = df.reset_index()
df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']

# Convert the 'Date' column to a datetime data type
df['Date'] = pd.to_datetime(df['Date'])

```

```python
df
```

```python
data=df.iloc[::-1].reset_index(drop = True)
```

```python
data.dtypes
```

```python
data['High'] = pd.to_numeric(data['High'])
data['Low'] = pd.to_numeric(data['Low'])
data['Open'] = pd.to_numeric(data['Open'])
data['Close'] = pd.to_numeric(data['Close'])
data['Volume'] = pd.to_numeric(data['Volume'])
```

```python
data.dtypes
```

```python
data.head()
```

```python
data['Total_Range'] = data['High'] - data['Low']
```

```python
data['Final_Close']=data['Close']-data['Open']
```

```python
data['Day Of Week'] = data.index.day_name()
```

```python
data['Positive_or_not'] = data.apply(lambda row: 'Positive' if (row['Open'] - row['Close']) < 0 else 'Negative', axis=1)
data['_%_Change'] = data.apply(lambda row: f"{round((row['Close'] - row['Open']) / row['Open'] * 100, 1)}%", axis=1)
```

```python
data['Day'] = data.index.day
data['Month'] = data.index.month
data['Year'] = data.index.year
```

```python
# data.set_index('Date', inplace=True)
```

## Get data from _._Every Year Data for Same Month_._

```python
# Assuming 'data' is your DataFrame, and 'target_month' is the month you want to analyze

target_month = 5

# Rest of the code remains unchanged
result_df = pd.DataFrame(columns=['Year', 'First_High', 'Last_Low', 'Month_range', 'Positive_or_not','_%_Change'])

for year in data['Year'].unique():
    target_data = data[(data.index.year == year) & (data.index.month == target_month)]

    if not target_data.empty:
        first_day_high = target_data.iloc[0]['Open']
        last_day_low = target_data.iloc[-1]['Close']

        result_df = result_df.append({
            'Year': year,
            'First_High': first_day_high,
            'Last_Low': last_day_low,
            'Month_range': last_day_low - first_day_high,
            'Positive_or_not': 'Positive' if (last_day_low - first_day_high) > 0 else 'Negative',
            '_%_Change': f"{round((last_day_low - first_day_high) / first_day_high * 100, 1)}%"
        }, ignore_index=True)

result_df
```

## graph on it 

```python
# Count the occurrences of each unique value in the 'plush_or_not' column
plush_counts = result_df['Positive_or_not'].value_counts()

# Define colors for positive and negative values
colors = ['red' if label == 'Negative' else 'green' for label in plush_counts.index]

# Plotting the bar plot with specified colors
plt.bar(plush_counts.index, plush_counts.values, color=colors)

# Adding labels and title
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('''Same Month for Every Year
Positive or Negative index''')
# Display the plot
plt.show()
```

# Year Base Report 

```python
data[data['Year'] == 2006]
```

```python
# Create a new DataFrame for year-based report
result_df = pd.DataFrame(columns=['Year', 'First_High', 'Last_Low', 'Year_range', 'Positive_or_not', '_%_Change'])

for year in data['Year'].unique():
    year_data = data[data['Year'] == year]

    if not year_data.empty:
        first_day_high = year_data.iloc[0]['Open']
        last_day_low = year_data.iloc[-1]['Close']

        result_df = result_df.append({
            'Year': year,
            'First_High': first_day_high,
            'Last_Low': last_day_low,
            'Year_range': last_day_low - first_day_high,
            'Positive_or_not': 'Positive' if (last_day_low - first_day_high) > 0 else 'Negative',
            '_%_Change': f"{round((last_day_low - first_day_high) / first_day_high * 100, 1)}%"
        }, ignore_index=True)

result_df = result_df.astype({'Year': int})



result_df
```

```python
# Count the occurrences of each unique value in the 'plush_or_not' column
plush_counts = result_df['Positive_or_not'].value_counts()

# Define colors for positive and negative values
colors = ['red' if label == 'Negative' else 'green' for label in plush_counts.index]

# Plotting the bar plot with specified colors
plt.bar(plush_counts.index, plush_counts.values, color=colors)

# Adding labels and title
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('''2005-2024 Market 
    Positive or Negative index''')
# Display the plot
plt.show()
```
