```python
import requests
from bs4 import BeautifulSoup
import json

url = "https://www.5paisa.com/stocks/all"

# Define headers to simulate a legitimate browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Make a GET request to the website with headers
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the div with class 'stock-directory__listname'
    stock_list_div = soup.find('div', class_='stock-directory__listname')

    # Find all ul tags within the div
    ul_tags = stock_list_div.find_all('ul')

    # Initialize a set to store unique stock names
    stock_names_set = set()

    # Iterate through ul tags and extract data from li tags
    for ul_tag in ul_tags:
        li_tags = ul_tag.find_all('li')
        for li_tag in li_tags:
            # Extract the text content from the <a> tag inside <li>
            stock_name = li_tag.find('a').text

            # Check if the stock name ends with 'Ltd'
            if stock_name.endswith('Ltd'):
                stock_names_set.add(stock_name)

    # Convert the set back to a list
    stock_names = list(stock_names_set)

    # Save the data in JSON format
    with open('stock_names.json', 'w') as json_file:
        json.dump(stock_names, json_file, indent=2)

    print("Unique stock names ending with 'Ltd' saved to stock_names.json")

else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.5paisa.com/stocks/all"

# Define headers to simulate a legitimate browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Make a GET request to the website with headers
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all links (a tags) within the page
    links = soup.find_all('a')

    # Initialize a set to store unique company names
    company_names_set = set()

    # Iterate through the links and extract the text content
    for link in links:
        company_name = link.text.strip()

        # Check if the company name ends with 'Ltd'
        if company_name.endswith('Ltd'):
            # Remove 'Ltd' and add to the set
            company_names_set.add(company_name[:-3].strip())

    # Convert the set back to a list
    company_names = list(company_names_set)
    
    # Save the data in JSON format    
    with open('stock_names.json', 'w') as json_file:
        json.dump(company_names, json_file, indent=2)

    print("Unique stock names ending with 'Ltd' saved to stock_names.json")
    
    # Print the modified company names
#     print("Company names without 'Ltd':", company_names)

else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```
