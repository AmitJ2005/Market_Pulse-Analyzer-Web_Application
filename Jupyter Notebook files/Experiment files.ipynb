```python
from yahooquery import search

# Replace 'COMPANY_NAME_TO_CHECK' with the specific company name you want to check
company_name_to_check = ' Tata Power Company'

try:
    # Use Yahoo Finance search to get stock symbols based on the company name
    result = search(company_name_to_check)

    if result['quotes']:
        # Take the first stock symbol from the search results
        stock_symbol = result['quotes'][0]['symbol']
        print(f"Stock symbol for {company_name_to_check}: {stock_symbol}")
    else:
        # Handle the case where no stock symbol is found
        print(f"No stock symbol found for {company_name_to_check}")
except Exception as e:
    print(f"Error fetching stock symbol for {company_name_to_check}: {e}")
```

```python
import yfinance as yf

# Define the stock symbol
symbol = "TATAPOWER.NS"

# Create a Ticker object for the stock
stock_ticker = yf.Ticker(symbol)

# Get company information
company_info = stock_ticker.info

# Extracting required information
general_info = {
    "Company Name": company_info.get("longName", ""),
    "Industry": company_info.get("industry", ""),
    "Website" : company_info.get("website", ""),
    "Total Employees": company_info.get("fullTimeEmployees"),
    'about' : company_info.get("longBusinessSummary", ""),
    "Close Price": company_info.get("previousClose", ""),
    "Market Capitalization": company_info.get("marketCap", ""),
    "52 Weaks low": company_info.get("fiftyTwoWeekLow",""),
    "52 Weaks high": company_info.get("fiftyTwoWeekHigh",""),
    "share in Market": company_info.get("floatShares",""),
    "Total Share in Company": company_info.get("sharesOutstanding",""),
    "Total Cash": company_info.get("totalCash"),
    "ebitda": company_info.get("ebitda",),
    "totalCashPerShare": company_info.get("totalCashPerShare"),
    "totalDebt": company_info.get("totalDebt",""),
    "totalRevenue": company_info.get("totalRevenue",""),
    "debtToEquity": company_info.get("debtToEquity",""),
    "revenuePerShare": company_info.get("revenuePerShare",""),
    "freeCashflow": company_info.get("freeCashflow",""),
    "grossMargins": company_info.get("grossMargins",""),
#     'about Company' : company_info.get("longBusinessSummary", "")
}

company_officers = []
for officer in company_info.get("companyOfficers", []):
    officer_info = {
        "Name": officer.get("name", ""),
        "Age": officer.get("age", ""),
        "Position": officer.get("title", ""),
        "Salary": officer.get("totalPay", "")
    }
    company_officers.append(officer_info)
```
